```python
import pandas as pd

# Page URL
url = 'https://www.basketball-reference.com/players/j/jordami01.html'

# Reading tables from page
tables = pd.read_html(url)

# Taking per game stats
per_game_stats = tables[0]

print(per_game_stats)
```

```python
per_game_stats.to_csv('michael_jordan_per_game_stats.csv', index=False)
```

```python
nba_data = pd.read_csv('NBA_Player_Stats.csv')

# Filter the dataset for only LeBron James
lebron_data = nba_data[nba_data['Player'] == 'LeBron James']

# Save the filtered dataset to a new file
lebron_data.to_csv('LeBron_stats.csv', index=False)

# Testing
print(lebron_data.head())
```
